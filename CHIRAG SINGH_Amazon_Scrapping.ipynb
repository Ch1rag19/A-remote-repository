```python
import requests
import pandas as pd
import os
from bs4 import BeautifulSoup
```

```python
review = []
names = []
date = []
loc = []
rating = []

for i in range(1,24):
    url = f'https://www.amazon.in/Apple-iPhone-11-Pro-64GB/product-reviews/B07XVM1JFS/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={i}'
    page = requests.get(url)
    if page.status_code ==200:
        soup = BeautifulSoup(page.text)
        # for names
        all_name = soup.find_all('span', {'class':'a-profile-name'})
        for ele in all_name[2:]:
            names.append(ele.text)
        # for loc and date
        all_date = soup.find_all('span', {'class': "a-size-base a-color-secondary review-date"})
        for element in all_date[2:]:
            temp = element.text.split(' on ')
            loc.append(temp[0])
            date.append(temp[1])
        # for rating
        all_rating = soup.find_all('span',{'class':'a-icon-alt'})
        for element in all_rating[3:]:
            rating.append(element.text.split(' out')[0])
        # for reviews
        all_review = soup.find_all('span',{'class':'a-size-base review-text review-text-content'})
        for element in all_review:
            element = element.text
            review.append(element[3:])
    else:
        print(f'Error Occured {page.status_code},on page no.{i}')
        
print('Reviews Scrapped Successfully..')
print('Creating Data Frame')
df = pd.DataFrame(data={
    'Name':names,
    'Date':date,
    'Location':loc,
    'Rating':rating,
    'Reviews':review
})
print('Data Frame Created')df.to_csv('amazon_review_details.csv')
```
